```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```
with open('Dosya.json', 'r') as dosya:
    veri=dosya.read()
```

```
obj = json.loads(veri)
```

```
ilkOkuma=False
for gun in obj:
  EnCokTercihEdilenler = gun['EnCokTercihEdilenler']
  EnAzTercihEdilenler = gun['EnAzTercihEdilenler']
  for cokTercih in EnCokTercihEdilenler:
    if ilkOkuma == False:
      yeni = [{f['firma']: 1 for f in EnAzTercihEdilenler}]
      df = pd.DataFrame(yeni, index=[cokTercih['firma']])
      ilkOkuma = True
    else:
      if cokTercih['firma'] in df.index.values:
        for f in EnAzTercihEdilenler:
          if f['firma'] in df.columns:
            v = df.at[cokTercih['firma'],f['firma']]
            df.at[cokTercih['firma'],f['firma']] = v + 1
          else:
            df.at[cokTercih['firma'],f['firma']]=1
      else:
        for f in EnAzTercihEdilenler:
          df.at[cokTercih['firma'],f['firma']] = 1
```

```
df = df.replace(np.nan,0)
```

```
plt.pcolor(df)
plt.yticks(np.arange(0.5, len(df.index), 1), df.index)
plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns)
fig = plt.gcf()
fig.set_size_inches(40, 10)
fig.savefig('Harita.png', dpi=100)
```
